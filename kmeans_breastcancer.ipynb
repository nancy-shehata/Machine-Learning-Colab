```python
from ast import increment_lineno
import numpy as np #linear alegbra
import pandas as pd #data processing, CSV file I/O (e.g. pd.read_csv)


import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
import matplotlib.cm as cm
%matplotlib inline


from sklearn import preprocessing
from subprocess import check_output
```

```python
#read the data file
data=pd.read_csv('Breastcancer.csv')
data.head()
```

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.manifold import TSNE

#Assuming 'data' is your DataFrame

#Check if 'id' column exists in DataFrame
if 'id' in data.columns:
  data=data.drop('id', axis=1)

#Mapping 'Benign' to 0 and 'Malignant' to 1

data['condition']= data['condition'].map({'M':1, 'B':0})

from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn import preprocessing
import pandas as pd
import matplotlib.pyplot as plt

#Assuming 'data' is a pre-loaded DataFrame with your data
#Scalling the dataset

datas=pd.DataFrame(preprocessing.scale(data.iloc[:,1:32]))
datas['condition']=data['condition']

#Creating the high dimesonal feature space x
data_drop=datas.drop('condition',axis=1)
X= data_drop.values

#Creating a 2D visiualisation to visualize the cluster
tsne=TSNE(verbose=1, perplexity=40, n_iter=4000)
#Perplexity- amount of inputs the algorthim will take each time it iterates
 #In the context of using t-SNE (t-Distributed Stochastic Neighbor Embedding) for dimensionality reduction and data visualization
Y= tsne.fit_transform(X) #This line should come before plotting
```

                        Notes:
More the algorthim iterates more the algorthim converge to a certain point (centroid)so that the similar data convergeo to the sinular centroids (central point)

```python
#CLuster using k-means
from sklearn.cluster import KMeans
kmns= KMeans(n_clusters=2,init='k-means++', n_init=10, max_iter=300, tol=0.0001,precompute_distances='auto', verbose=0, random_state=None, copy_x=True, n_jobs=1, algorithm='auto')
kY=kmns.fit_predict(X)

f,(ax1,ax2)= plt.subplots(1,2, sharey=True)

ax1.scatter(Y[:,0],Y[:,1], c=kY, cmap="jet", edgecolor="None", alpha=0.35)
ax1.set_title('k-means cluster plot')

ax2.scatter(Y[:,0], Y[:,1], c=datas['condition'], cmap="jet", edgecolor="None", alpha=0.35)
ax2.se_title('Actual clusters')
```
